```python
UK=Q4[Q4['CountryName']=='United Kingdom']
UKSUM=UK.sum(axis=0, numeric_only=True)
UK_SALES=round(UKSUM['SalePrice'])
UK_COST=round(UKSUM['CostPrice'])
```

```python
US=Q4[Q4['CountryName']=='USA']
USSUM=US.sum(axis=0, numeric_only=True)
US_SALES=round(USSUM['SalePrice'])
US_COST=round(USSUM['CostPrice'])
```

```python
FRANCE=Q4[Q4['CountryName']=='France']
FRANCESUM=FRANCE.sum(axis=0, numeric_only=True)
FRANCE_SALES=round(FRANCESUM['SalePrice'])
FRANCE_COST=round(FRANCESUM['CostPrice'])
```

```python
Combined=pd.DataFrame([('Usa',US_SALES,US_COST),
                    ('Uk',UK_SALES,UK_COST),
                    ('France',FRANCE_SALES,FRANCE_COST)],
                   columns=['Country','Total Sales',"Total Cost"])
```

```python
Combined.plot('Country',['Total Sales','Total Cost'], kind='barh', )
```
