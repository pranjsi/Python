```python
df_dade = df_all[df_all['County'].isin(['Dade'])]
```

```python
df_dade
```

```python
# Group the data by county name 'Dade' using count and sum.

df_dade_count = df_dade.groupby(['EventDate']).count() 
df_dade_count.index = pd.to_datetime(df_dade_count.index, format="%m/%d/%Y")
df_dade_count_sorted = df_dade_count.sort_index(axis='index', ascending=True)
df_dade_count_sorted['Date'] = df_dade_count_sorted.index.values
df_dade_count_sorted['Daily Cases'] = df_dade_count_sorted['County']
```

```python
df_dade_sum = df_dade.groupby(['EventDate']).sum() 
df_dade_sum.index = pd.to_datetime(df_dade_sum.index, format="%m/%d/%Y")
df_dade_sum_sorted = df_dade_sum.sort_index(axis='index', ascending=True)
df_dade_sum_sorted['Date'] = df_dade_sum_sorted.index.values
```

```python
df_Orange = df_all[df_all['County'].isin(['Orange'])]
```

```python
df_Orange
```

```python
# Group the data by county name 'Orange' by count and Sum

df_Orange_count = df_Orange.groupby(['EventDate']).count() 
df_Orange_count.index = pd.to_datetime(df_Orange_count.index, format="%m/%d/%Y")
df_Orange_count_sorted = df_Orange_count.sort_index(axis='index', ascending=True)
df_Orange_count_sorted['Date'] = df_Orange_count_sorted.index.values
df_Orange_count_sorted['Daily Cases'] = df_Orange_count_sorted['County']
```

```python
df_Orange_sum = df_Orange.groupby(['EventDate']).sum() 
df_Orange_sum.index = pd.to_datetime(df_Orange_sum.index, format="%m/%d/%Y")
df_Orange_sum_sorted = df_Orange_sum.sort_index(axis='index', ascending=True)
df_Orange_sum_sorted['Date'] = df_Orange_sum_sorted.index.values
```

```python
Orange = df_Orange_count_sorted
Dade = df_dade_count_sorted
```

```python
Orange['Moving Average'] = Orange['Daily Cases'].rolling(5).mean()
Dade['Moving Average'] = Dade['Daily Cases'].rolling(5).mean()
```

```python
fig, h_axes = mp.subplots(2,2,figsize=[10,10])

fig.suptitle('Daily Cases and Moving average Dashboard for Dade and Orange County')

mp.sca(h_axes[0,0])
mp.plot(Dade['Date'], Dade['Daily Cases'], c='black')
h_axes[0,0].set_title('Daily Number of Cases for Dade County')
h_axes[0,0].set_xlabel('Date')
h_axes[0,0].set_xticks([])

mp.sca(h_axes[0,1])
mp.plot(Dade['Date'], Dade['Moving Average'],c='black')
h_axes[0,1].set_title('Moving Average of Cases for Dade County')
h_axes[0,1].set_xlabel('Date')
h_axes[0,1].set_xticks([])

mp.sca(h_axes[1,0])
mp.plot(Orange['Date'], Orange['Daily Cases'], c='orange')
h_axes[1,0].set_title('Daily Number of Cases for Orange County')
h_axes[1,0].set_xlabel('Date')
h_axes[1,0].set_xticks([])

mp.sca(h_axes[1,1])
mp.plot(Orange['Date'], Orange['Moving Average'], c='orange')
h_axes[1,1].set_title('Moving Average of Cases for Orange County')
h_axes[1,1].set_xlabel('Date')
h_axes[1,1].set_xticks([])


h_axes[0,0].set_ylim(bottom= 0, top=4500)
h_axes[0,1].set_ylim(bottom= 0, top=4500)

h_axes[1,0].set_ylim(bottom= 0, top=1400)
h_axes[1,1].set_ylim(bottom= 0, top=1400)
```
